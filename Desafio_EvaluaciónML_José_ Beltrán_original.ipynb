```python
# Llibrerías a utilizar
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
mpl.style.use('ggplot')
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
from sklearn.model_selection import TimeSeriesSplit
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
```

```python
# Importamos el dataset
vgamesales_url = 'https://raw.githubusercontent.com/jbeltrang/datasets-ch/main/Video_Games_Sales_as_at_22_Dec_2016.csv'
vgamesales = pd.read_csv(vgamesales_url)
vgamesales

# Hacemos un dataframe trabajable y limpiamos la data que sea mayor al 2016
vgamesales_df = vgamesales
vgamesales_df = vgamesales_df.loc[vgamesales_df['Year_of_Release'] <= 2016]
vgamesales_df = vgamesales_df.dropna(subset=['Year_of_Release'])
```

## Selección de Variables y Encoding

Para el caso que debemos resolver, supondremos que el género del videojuego a desarrollar en el proyecto corresponde al género de Acción. Dicho esto, procedemos a realizar la selección de variables.

```python
# Definimos un subset para trabajar la selección de variables

vgamesales_set = vgamesales
# Drop de valores nulos y limpieza
vgamesales_set = vgamesales_set.dropna(subset=['Year_of_Release','Genre',"Critic_Score","User_Score","NA_Sales","EU_Sales","JP_Sales","Other_Sales"])
vgamesales_set = vgamesales_set.loc[vgamesales_set['User_Score'] != "tbd"]
# Conversión de fechas
vgamesales_set['Year_of_Release'] = pd.to_datetime(vgamesales_set['Year_of_Release'], format="%Y")
vgamesales_set['Year'] = vgamesales_set['Year_of_Release'].dt.year
# Obtener la columna de años
years = vgamesales_set['Year']
# Calcular las décadas correspondientes
decades = (years // 10) * 10
# Subset
vgamesales_set = vgamesales_set[["Platform","Year","Genre","Global_Sales","Critic_Score","User_Score","NA_Sales","EU_Sales","JP_Sales","Other_Sales"]]
# One-hot Encoding
vgamesales_set = pd.get_dummies(vgamesales_set, columns=["Platform","Genre"])
decades_encoded = pd.get_dummies(decades, prefix='Decade')
# Concatenar las columnas de décadas con el dataset original
vgamesales_set_encoded = pd.concat([vgamesales_set, decades_encoded], axis=1)
# Eliminar la columna original de "Year"
vgamesales_set_encoded.drop('Year', axis=1, inplace=True)
vgamesales_set = vgamesales_set_encoded

# Normalización de los Scores
Critic_Score_normalizado = StandardScaler().fit_transform(vgamesales_set['Critic_Score'].values.reshape(-1, 1))
vgamesales_set['Critic_Score'] = Critic_Score_normalizado
User_Score_normalizado = StandardScaler().fit_transform(vgamesales_set['User_Score'].values.reshape(-1, 1))
vgamesales_set['User_Score'] = User_Score_normalizado

# Normalización de Sales
User_Score_normalizado = StandardScaler().fit_transform(vgamesales_set['NA_Sales'].values.reshape(-1, 1))
vgamesales_set['NA_Sales'] = User_Score_normalizado
User_Score_normalizado = StandardScaler().fit_transform(vgamesales_set['EU_Sales'].values.reshape(-1, 1))
vgamesales_set['EU_Sales'] = User_Score_normalizado
User_Score_normalizado = StandardScaler().fit_transform(vgamesales_set['EU_Sales'].values.reshape(-1, 1))
vgamesales_set['EU_Sales'] = User_Score_normalizado
User_Score_normalizado = StandardScaler().fit_transform(vgamesales_set['JP_Sales'].values.reshape(-1, 1))
vgamesales_set['JP_Sales'] = User_Score_normalizado
User_Score_normalizado = StandardScaler().fit_transform(vgamesales_set['Other_Sales'].values.reshape(-1, 1))
vgamesales_set['Other_Sales'] = User_Score_normalizado

vgamesales_set
```

## Análisis de Correlación

## Modelamiento

```python
X = vgamesales_set.drop('Global_Sales', axis=1)
y = vgamesales_set['Global_Sales']

# Dividir el dataset en conjuntos de entrenamiento y prueba
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Crear y entrenar el modelo de regresión lineal
model = LinearRegression()
model.fit(X_train, y_train)

# Realizar predicciones en el conjunto de prueba
y_pred = model.predict(X_test)

# Evaluar el rendimiento del modelo
# Calcular el coeficiente de determinación (R2)
r2 = r2_score(y_test, y_pred)
print("Coeficiente de determinación (R2):", r2)

# Calcular el error cuadrático medio (MSE)
mse = mean_squared_error(y_test, y_pred)
print("Error cuadrático medio (MSE):", mse)

# Calcular la raíz del error cuadrático medio (RMSE)
rmse = mean_squared_error(y_test, y_pred, squared=False)
print("Raíz del error cuadrático medio (RMSE):", rmse)

# Calcular el error absoluto medio (MAE)
mae = mean_absolute_error(y_test, y_pred)
print("Error absoluto medio (MAE):", mae)
```

## Interpretación

**Coeficiente de determinación (R2)**

Este valor, que está cerca de 1 (0.99), indica que el modelo explica casi la totalidad de la variabilidad de los datos de manera muy precisa. En otras palabras, el 99.99% de la variación en la variable dependiente puede ser explicada por las variables independientes del modelo.

**Error cuadrático medio (MSE)**

El valor de MSE obtenido es 3.43 indica que los errores de predicción son muy pequeños en promedio, lo que sugiere un buen ajuste del modelo a los datos.

**Raíz del error cuadrático medio (RMSE)**

El valor de RMSE obtenido es 0.006, que es la raíz cuadrada del MSE. Al igual que el MSE, cuanto menor sea el valor de RMSE, mejor será la precisión del modelo. En este caso, un RMSE de 0.006 indica que, en promedio, las predicciones difieren en aproximadamente 0.006 unidades de los valores reales. Si lo pensamos en términos de números, considerando que el valor mayor es 98 y el menor 0,01, las desviaciones serían mínimas, dando una alta cercanía a los valores reales.

**Error absoluto medio (MAE)**

El valor de MAE obtenido es 0.004, que es la media de los errores absolutos entre las predicciones y los valores reales. En este caso, un MAE de 0.04 indica que, en promedio, las predicciones difieren en aproximadamente 0.04 unidades de los valores reales, lo cual indica que las predicciones del modelo tienden a estar muy cerca de los valores reales en promedio.

## Validación Cruzada

```python
# Realizar validación cruzada con 20 folds
scores = cross_val_score(model, X, y, cv=20)

# Mostrar los resultados
print("Puntajes de validación cruzada:", scores)
print("Puntaje promedio:", scores.mean())
```

## Interpretación

El incluir las ventas de las distintas regiones comerciales de forma estandarizada parece haber tenido un efecto bastante positivo en el modelo. También es necesario considerar el aumento del conjunto de prueba en un 30%, lo cual puede haber influído en cierta medida.

El puntaje promedio de validación cruzada es de 0.6852, lo cual indica un rendimiento moderado del modelo. Este puntaje promedio sugiere que el modelo puede tener dificultades para generalizar correctamente a partir de los datos de entrenamiento a los datos de validación. Puede haber cierto grado de sobreajuste o falta de capacidad para capturar los patrones importantes en los datos.
